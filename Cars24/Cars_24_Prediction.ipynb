```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
data['Fuel']=le.fit_transform(data['Fuel'])
print(le.classes_)
data['Location']=le.fit_transform(data['Location'])
print(le.classes_)
data['Drive']=le.fit_transform(data['Drive'])
print(le.classes_)
data['Type']=le.fit_transform(data['Type'])
print(le.classes_)
data['Car_Name']=le.fit_transform(data['Car_Name'])
print(le.classes_)
```

```python
data
```

```python
X=data.drop('Price',axis=1)
y=data['Price']
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

### The models that can be tried are:

1. Linear Regression
2. Polynomial Regression
3. SGD Regression
4. Lasso Regression
5. Ridge Regression
6. KNN Regressor
7. Ensemble
8. Bagging and Boosting

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV

polynomial_regression_pipeline = Pipeline([
    ("poly", PolynomialFeatures(interaction_only=True)),
    ("Linear", LinearRegression())
    ])

param_grid = {'poly__degree': (1,2,3,4)}

polynomial_regression = GridSearchCV(estimator= polynomial_regression_pipeline,
                                param_grid = param_grid,
                                scoring = "r2",
                                return_train_score=True,
                                cv = 5
                               )
polynomial_regression.fit(X_train,y_train)

cv_results_df = pd.DataFrame(polynomial_regression.cv_results_)
print(polynomial_regression.best_params_)
print(polynomial_regression.best_score_)
```

```python
from sklearn.linear_model import Ridge
from sklearn.preprocessing import StandardScaler

ridge_estimator = Ridge(alpha=1)
param_grid={
    'alpha':[1e-15,1e-10,1e-8,1e-4,1e-3,1e-2,1,5,10,20],
    'solver':['svd','cholesky','sag','saga']
}


ridge_regression=GridSearchCV(ridge_estimator,
                              param_grid=param_grid,
                              n_jobs=2,
                              cv=5,
                              scoring='r2',
                              return_train_score=True)
ridge_regression.fit(X_train,y_train)
```

```python
ridge_regression.best_params_
```

```python
ridge_regression.best_score_
```

```python
from sklearn.linear_model import Lasso
from sklearn.preprocessing import StandardScaler

lasso_estimator = Lasso(alpha=1,warm_start=True,random_state=42)
param_grid={
    'alpha':[1e-15,1e-10,1e-8,1e-4,1e-3,1e-2,1,5,10,20],
    'max_iter':[100,250,500,1000,2000,5000],
    'tol':[1e-15,1e-12,1e-10,1e-5,1e-2,1e-1,1,10,100,1000],
    'selection':['cyclic','random']
}


lasso_regression=GridSearchCV(lasso_estimator,
                              param_grid=param_grid,
                              cv=5,
                              scoring='r2',
                              return_train_score=True)
lasso_regression.fit(X_train,y_train)
```

```python
lasso_regression.best_params_
```

```python
lasso_regression.best_score_
```

```python
from sklearn.ensemble import RandomForestRegressor

rfg_reg=RandomForestRegressor(random_state=42)
rfg_reg.fit(X_train,y_train)
y_pred_rfg=rfg_reg.predict(X_test)
r2_score(y_test,y_pred_rfg)


randomforestestimator=RandomForestRegressor(bootstrap=True,random_state=42)
param_grid={
    'n_estimators':[50,100],
    'criterion':['squared_error','poisson']
}


RFC_regression=GridSearchCV(randomforestestimator,
                              param_grid=param_grid,
                              cv=5,
                              scoring='r2',
                              return_train_score=True)
RFC_regression.fit(X_train,y_train)
```
